```python
def sum_even_odd(numbers):
  """Calculates the sum of even and odd numbers in a list.

  Args:
    numbers: A list of integers.

  Returns:
    A tuple containing the sum of even numbers and the sum of odd numbers.
    (sum_of_evens, sum_of_odds).

  Example:
    >>> sum_even_odd([1, 2, 3, 4, 5, 6])
    (12, 9)
  """
  sum_evens = 0  # Initialize a variable to store the sum of even numbers
  sum_odds = 0   # Initialize a variable to store the sum of odd numbers

  for number in numbers:  # Iterate through each number in the input list
    if number % 2 == 0:   # Check if the current number is even using the modulo operator
      sum_evens += number # Add the even number to the sum of evens
    else:                 # If the number is not even, it must be odd
      sum_odds += number  # Add the odd number to the sum of odds

  return (sum_evens, sum_odds) # Return the tuple containing the sums of evens and odds

# Example usage
my_list = [1, 2, 3, 4, 5, 6]
even_sum, odd_sum = sum_even_odd(my_list)

print(f"The list is: {my_list}")
print(f"Sum of even numbers: {even_sum}")
print(f"Sum of odd numbers: {odd_sum}")
```

```python
class sru_student:
    """
    Represents a student at SRU.
    """

    def __init__(self, name, roll_no, hostel_status):
        """
        Initializes a new sru_student object.

        Args:
            name (str): The name of the student.
            roll_no (str): The roll number of the student.
            hostel_status (bool): True if the student is a hosteller, False otherwise.
        """
        self.name = name  # Store name
        self.roll_no = roll_no  # Store roll number
        self.hostel_status = hostel_status  # Store hostel status
        self.fees_paid = 0  # Initialize fees paid

    def fee_update(self, amount):
        """
        Updates the fees paid by the student.

        Args:
            amount (float): The amount of fee payment.

        Raises:
            ValueError: If the amount is negative.
        """
        if amount < 0: # Check if amount is negative
            raise ValueError("Fee amount cannot be negative.") # Raise error if negative
        self.fees_paid += amount  # Add amount to fees paid

    def display_details(self):
        """
        Displays the details of the student.
        """
        print("Student Details:") # Print header
        print(f"Name: {self.name}")  # Print name
        print(f"Roll No: {self.roll_no}")  # Print roll number
        print(f"Hostel Status: {'Hosteller' if self.hostel_status else 'Day Scholar'}") # Print hostel status
        print(f"Fees Paid: {self.fees_paid}")  # Print fees paid


# Example Usage:
# Create student object
student1 = sru_student("Alice Smith", "SRU123", True)

# Display initial details
student1.display_details()

print("\nUpdating fees...")

# Update fees
student1.fee_update(5000.0)

# Display details after update
student1.display_details()

print("\nAttempting to update with a negative amount...")

# Attempt negative update for error handling
try:
    student1.fee_update(-1000.0)
except ValueError as e:
    print(f"Error: {e}") # Print error message

# Display details again
student1.display_details()
```

```python
"""
Arithmetic module. Provides basic operations.
"""

def add(a, b):
    """Adds two numbers.

    Parameters
    ----------
    a : numeric
        First number.
    b : numeric
        Second number.

    Returns
    -------
    numeric
        Sum of a and b.
    """
    return a + b

def subtract(a, b):
    """Subtracts b from a.

    Parameters
    ----------
    a : numeric
        Minuend.
    b : numeric
        Subtrahend.

    Returns
    -------
    numeric
        Difference of a and b.
    """
    return a - b

def multiply(a, b):
    """Multiplies two numbers.

    Parameters
    ----------
    a : numeric
        First number.
    b : numeric
        Second number.

    Returns
    -------
    numeric
        Product of a and b.
    """
    return a * b

def divide(a, b):
    """Divides a by b.

    Parameters
    ----------
    a : numeric
        Dividend.
    b : numeric
        Divisor.

    Returns
    -------
    float
        Result of a / b.

    Raises
    ------
    ZeroDivisionError
        If b is zero.
    """
    if b == 0:
        # Check for division by zero to prevent an error.
        raise ZeroDivisionError("Division by zero is not allowed.")
    return a / b

# Usage Examples (same as before for comparison)

# Example for add
num1 = 10
num2 = 5
sum_result = add(num1, num2)
print(f"Addition: {num1} + {num2} = {sum_result}")

# Example for subtract
diff_result = subtract(num1, num2)
print(f"Subtraction: {num1} - {num2} = {diff_result}")

# Example for multiply
prod_result = multiply(num1, num2)
print(f"Multiply: {num1} * {num2} = {prod_result}")

# Example for divide
div_result = divide(num1, num2)
print(f"Divide: {num1} / {num2} = {div_result}")

# Example for divide with zero
try:
    divide(num1, 0)
except ZeroDivisionError as e:
    print(f"Error during division: {e}")
```

```python
"""
This is a simple arithmetic module.

It provides basic functions for addition, subtraction, multiplication,
and division of two numbers.
"""

def add(a, b):
    """
    Adds two numbers.

    Parameters
    ----------
    a : numeric
        The first number.
    b : numeric
        The second number.

    Returns
    -------
    numeric
        The sum of a and b.
    """
    return a + b

def subtract(a, b):
    """
    Subtracts the second number from the first.

    Parameters
    ----------
    a : numeric
        The first number (minuend).
    b : numeric
        The second number (subtrahend).

    Returns
    -------
    numeric
        The difference between a and b.
    """
    return a - b

def multiply(a, b):
    """
    Multiplies two numbers.

    Parameters
    ----------
    a : numeric
        The first number.
    b : numeric
        The second number.

    Returns
    -------
    numeric
        The product of a and b.
    """
    return a * b

def divide(a, b):
    """
    Divides the first number by the second.

    Parameters
    ----------
    a : numeric
        The dividend.
    b : numeric
        The divisor.

    Returns
    -------
    float
        The result of the division a / b.

    Raises
    ------
    ZeroDivisionError
        If the divisor b is zero.
    """
    if b == 0:
        # Manual comment: Check for division by zero to prevent an error.
        raise ZeroDivisionError("Division by zero is not allowed.")
    return a / b

# Usage Examples:

# Example for add
num1 = 10
num2 = 5
sum_result = add(num1, num2)
print(f"Addition: {num1} + {num2} = {sum_result}") # Expected output: Addition: 10 + 5 = 15

# Example for subtract
diff_result = subtract(num1, num2)
print(f"Subtraction: {num1} - {num2} = {diff_result}") # Expected output: Subtraction: 10 - 5 = 5

# Example for multiply
prod_result = multiply(num1, num2)
print(f"Multiply: {num1} * {num2} = {prod_result}") # Expected output: Multiply: 10 * 5 = 50

# Example for divide
div_result = divide(num1, num2)
print(f"Divide: {num1} / {num2} = {div_result}") # Expected output: Divide: 10 / 5 = 2.0

# Example for divide with zero (to show Raises in docstring)
try:
    divide(num1, 0)
except ZeroDivisionError as e:
    print(f"Error during division: {e}") # Expected output: Error during division: Division by zero is not allowed.
```
